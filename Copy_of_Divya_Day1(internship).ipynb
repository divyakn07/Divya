```python
#nested if-else
#TO-DO
#add all the comments
i=40
if(i==40):"""this condition checks if i is equasls to 40 or not.if the condition holds true then it goes into the loop"""
  if(i<=50):#here it checks wether the value i is lesser than or equal to 15 if the condition holds true then it prints "i is smaller than 15"
    print("i is smaller than 50")
  if(i<12):"""here it checks wether the value of i is lesser than 12 if the condition holds true then it prints "i is smaller than 12 too",
               if the condition holds false it print the statement in else block"""
    print("i is smaller than 12 too")

  else:
    print("i is greater than 15")

else:
  print("i is not 40")#if the value of i is not 40 then it directly executes this else block saying "i is not 40"
```
